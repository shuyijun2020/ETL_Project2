```python
# import Dependencies
from bs4 import BeautifulSoup
import requests
import pandas as pd
import re
from sqlalchemy import create_engine
```

```python
# create a function that can be reused for scraping the data
def get_html_data(url):
    response = requests.get(url)
    return BeautifulSoup(response.text, "html5lib")
```

```python
hfm_url = 'https://www.pro-football-reference.com/hof/index.htm'

hfm_response = get_html_data(hfm_url)
print(hfm_response.prettify())
```

```python
# results are returned as an iterable list
hfm_results = hfm_response.find_all('table', class_="sortable stats_table")

#print(hfm_results)
print(hfm_results[0].prettify())   

hfm_df = pd.read_html(str(hfm_results[0])
hfm_df = hfm_df[0]
hfm_df.head()
```
